```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv('CASP.csv')  # Убедитесь, что файл находится в рабочей директории

# Вычисление матрицы корреляций
correlation_matrix = df.corr()

# Настройка размера графика
plt.figure(figsize=(12, 10))

# Создание тепловой карты
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            linewidths=0.5,
            annot_kws={"size": 8})

# Настройка отображения
plt.title('Матрица корреляций для датасета Protein Tertiary Structure', pad=20, fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
